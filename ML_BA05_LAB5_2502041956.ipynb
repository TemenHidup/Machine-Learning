```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel('dataset_lab5_1.xlsx')
df2 = pd.read_excel('dataset_lab5_2.xlsx')

print(df1.head())
print(df2.head())
```

```python
# 1

from sklearn.cluster import KMeans

inertia = []
k = range(1,10)
for i in k:
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(df1)
  y_kmeans = kmeans.predict(df1)
  inertia.append(kmeans.inertia_)
```

```python
plt.plot(k, inertia)
plt.xlabel('K')
plt.ylabel('Inertia')
plt.show()
print("the most effective “K” is 4 ")
```

```python
#2. 
kmeans = KMeans(n_clusters = 4)
kmeans.fit(df1)
y_kmeans = kmeans.predict(df1)

fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2,2)
fig.set_figwidth(20)
fig.set_figheight(10)
sns.scatterplot(ax = ax1, x=df1['A'], y=df1['B'], hue = y_kmeans)
sns.scatterplot(ax = ax2, x=df1['B'], y=df1['C'], hue = y_kmeans)
sns.scatterplot(ax = ax3, x=df1['C'], y=df1['D'], hue = y_kmeans)
sns.scatterplot(ax = ax4, x=df1['D'], y=df1['A'], hue = y_kmeans)
ax1.set_title('A VS B')
ax1.set_xlabel('A')
ax1.set_ylabel('B')
ax2.set_title('B VS C')
ax2.set_xlabel('B')
ax2.set_ylabel('C')
ax3.set_title('C VS D')
ax3.set_xlabel('C')
ax3.set_ylabel('D')
ax4.set_title('D VS A')
ax4.set_xlabel('D')
ax4.set_ylabel('A')

plt.tight_layout()
```

```python
#3.

from sklearn.mixture import GaussianMixture
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.3).fit(df2)
y_dbscan = dbscan.labels_

kmeans = KMeans(n_clusters = 5)
kmeans.fit(df2)
y_kmeans = kmeans.predict(df2)

gaussian = GaussianMixture(n_components = 5)
gaussian.fit(df2)
y_gaussian = gaussian.predict(df2)

fig, (ax1, ax2,ax3) = plt.subplots(1,3)
fig.set_figwidth(20)
fig.set_figheight(10)
sns.scatterplot(ax=ax1, x=df2['A'], y=df2['B'], hue = y_kmeans)
sns.scatterplot(ax=ax2, x=df2['A'], y=df2['B'], hue = y_gaussian)
sns.scatterplot(ax=ax3, x=df2['A'], y=df2['B'], hue = y_dbscan)
ax1.set_title('KMeans')
ax2.set_title('Gaussian')
ax3.set_title('DBScan')
```
